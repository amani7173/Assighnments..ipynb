```python
# Use Values to return a Nummpy array to filter our data with
# Documentation specifically asks for an array

# Return all rows where "deck" is NaN
df.iloc[df["deck"].isna().values, :]
```

```python
# Return all rows where "deck" is NOT a NaN
df.iloc[df["deck"].isna().values == False, : ]
```

```python
# You can also cast as a list to satisfy the requirements of teh boolean array
df.iloc[list(df["deck"].isna() == False), :]
```

### Pivot Tables

[Reference](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html)

```python
help(pd.pivot_table)
```

```python
# Values represent the perentage that survuved in each class for each gender

df.pivot_table(values = "survived", 
               index="sex", 
               columns="class",
              )
```

```python
df.pivot_table(values = "survived", 
               index="sex", 
               columns="class",
               aggfunc="count"
              )
```

```python
import numpy as np

df.pivot_table(index='sex',
               columns='class',
               aggfunc={'survived':sum, 'fare': np.mean})
```

```python
# Joining data with an implicit inner join

left = pd.DataFrame(
     {
         "key1": ["K0", "K1", "K2", "K3"],
         "A": ["A0", "A1", "A2", "A3"],
         "B": ["B0", "B1", "B2", "B3"],
     }
 )


right = pd.DataFrame(
     {
         "key1": ["K0", "K1", "K2", "K3"],
         "C": ["C0", "C1", "C2", "C3"],
         "D": ["D0", "D1", "D2", "D3"],
     }
 )
 

result = pd.merge(left, right, on="key1", how="inner")

result
```

```python
left.merge(right, on="key1", how="inner")
```

```python
# Joining data on multiple keys

# Add a new key to each of the dataframes
left["key2"] = ["K0", "K1", "K0", "K1"]
right["key2"] = ["K0", "K0", "K0", "K0"]

result = pd.merge(left, right, on=["key1", "key2"], how="outer")
result
```
